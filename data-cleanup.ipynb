```python
# import libraries

from cenpy import products
import geopandas as gpd
import pandas as pd
```

# American Community Survey Data

```python
# American Community Survey
acs = products.ACS(2019)
```

```python
# Find tables containing 'INTERNET'
acs.filter_tables('INTERNET', by='description')
```

```python
# Print list of tables
acs.filter_variables('B28011')
```

```python
# Download data
or_acs = products.ACS(2019).from_state('Oregon', 
                                          level='county',
                                          variables=['B28011'])
or_acs_reprojected = or_acs.to_crs('EPSG:32610')
```

```python
# convert ACS data to DataFrame
or_acs_df = pd.DataFrame(or_acs_reprojected)

# save DataFrame to file
or_acs_df.to_csv('../Data/or_acs.csv', index=False)
```

# Ookla Internet Speed Data

```python
# improt internet speed tiles
tiles = gpd.read_file('../Data/2022-07-01_performance_fixed_tiles/gps_fixed_tiles.shp')
```

```python
tiles_reprojected = tiles.to_crs('EPSG:32610')
```

```python
or_tiles = gpd.sjoin(tiles_reprojected, or_acs_reprojected, how='inner', op='within')
or_tiles.head()
```

```python
# save GeoDataFrame to file
or_tiles.to_file('../Data/or_tiles.shp', drive='ESRI Shapefile')
```
