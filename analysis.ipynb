```python
# import libraries

import geopandas as gpd
import pandas as pd
```

```python
or_tiles = gpd.read_file('../Data/or_tiles.shp')
```

```python
or_tiles.head()
```

```python
or_counties = or_tiles.dissolve(by='NAME', aggfunc='mean') 
```

```python
or_counties
```

```python
# convert speeds to mbps
or_counties['avg_d_mbps'] = or_counties['avg_d_kbps'] / 1000
or_counties['avg_u_mbps'] = or_counties['avg_u_kbps'] / 1000
```
