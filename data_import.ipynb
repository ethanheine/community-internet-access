```python
# import libraries

from datetime import datetime
from cenpy import products
import geopandas as gpd
import pandas as pd
```

# American Community Survey Data

```python
# American Community Survey
acs = products.ACS(2019)
```

```python
# Find tables containing 'INTERNET'
acs.filter_tables('INTERNET', by='description')
```

```python
# Print list of tables
acs.filter_variables('B28011')
```

```python
# Download data

# Oregon counties
or_acs_counties = products.ACS(2019).from_state('Oregon', 
                                          level='county',
                                          variables=['B28011'])

# Oregon tracts
or_acs_tracts = products.ACS(2019).from_state('Oregon',
                                        level='tract',
                                        variables=['B28011'])

# Reproject data
counties_reprojected = or_acs_counties.to_crs('EPSG:32610')
tracts_reprojected = or_acs_tracts.to_crs('EPSG:32610')
```

```python
# save data to folder
counties_reprojected.to_file('../Data/or_counties_acs.shp')
tracts_reprojected.to_file('../Data/or_tracts_acs.shp')
```

# Ookla Internet Speed Data

```python
# Ookla code to download data
def quarter_start(year: int, q: int) -> datetime:
    if not 1 <= q <= 4:
        raise ValueError("Quarter must be within [1, 2, 3, 4]")

    month = [1, 4, 7, 10]
    return datetime(year, month[q - 1], 1)


def get_tile_url(service_type: str, year: int, q: int) -> str:
    dt = quarter_start(year, q)

    base_url = "https://ookla-open-data.s3-us-west-2.amazonaws.com/shapefiles/performance"
    url = f"{base_url}/type%3D{service_type}/year%3D{dt:%Y}/quarter%3D{q}/{dt:%Y-%m-%d}_performance_{service_type}_tiles.zip"
    return url
```

```python
tile_url = get_tile_url('fixed', 2020, 2)
tile_url
```

```python
# Download Ookla speed test data and reproject
tiles = gpd.read_file(tile_url)
tiles_reprojected = tiles.to_crs('EPSG:32610')
```

```python
# merge ACS and Ookla GeoDataFrames
or_tract_tiles = gpd.sjoin(tiles_reprojected, tracts_reprojected, how='inner', op='within')

or_tract_tiles.head()
```

```python
# save GeoDataFrame to file
or_tract_tiles.to_file('../Data/or_tract_tiles.shp', drive='ESRI Shapefile')
```
